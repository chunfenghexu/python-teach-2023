# 0-1背包问题
问题描述：
>我们有一个背包，背包总的承载重量是 Wkg。现在我们有 n 个物品，每个物品的重量不等，并且不可分割。我们现在期望选择几件物品，装载到背包中。在不超过背包所能装载重量的前提下，如何让背包中物品的总重量最大？物品是不可分割的，要么装要么不装，所以叫 0-1 背包问题

```python
def backpack01(w,n,c):
    '''
    w:list，物品重量的列表
    n:物品数量
    c:背包容量
    '''
    # curc[i][j] : i表示该决策第几个物品了；j表示当前背包的重量；True of False，表示是否存在这个状态
    # curc[0][0] : 当对第0个物品进行决策时，此时背包还什么也没有装
    curc = [[False for j in range(c+1)] for i in range(n)]
    # 第0个物品装或不装
    curc[0][0] = True
    if c >= w[0]:
        curc[0][w[0]] = True
    for i in range(1,n):
        # 对第i个物品进行决策
        for j in range(c+1): # 不装
            if curc[i-1][j]:curc[i][j]=True
        for j in range(c+1): # 装
            if curc[i-1][j] and j+w[i]<=c:curc[i][j+w[i]]=True # 当装完第i个的时候不能超出背包容量
    # 最终，curc中所有为True的值都是可能存在的情况，我们只需要在最后一个物品决策完成后，找到背包当前容量最多那个即可
    print_backpack01_curcMatrix(w,curc,n,c+1)
    get_backpack01_path(w,curc,n,c)
    for k in range(c,-1,-1):
        if curc[n-1][k]==True:break
    return (k,curc)
def print_backpack01_curcMatrix(w,curc,rown,coln):
    print("\\",end="   ")
    for j in range(coln):
        print(j,end=" ")
    print("")
    for j in range(coln):
        print("-",end="-")
    print("---")
    for i in range(rown):
        print("{},{}".format(i,w[i]),end=":")
        for j in range(coln):
            print(1 if curc[i][j] else 0,end=" ")
        print("")
def get_backpack01_path(w,curc,rown,coln):
    for k in range(coln,-1,-1):
        if curc[rown-1][k]==True:break
    maxS = k
    maxS_ = 0
    path = [0 for i in range(rown)]
    for i in range(rown-1,-1,-1):#倒着往前回看
        # 要么装了，要么没装
        if k-w[i] >= 0 and curc[i-1][k-w[i]]==True:
            k = k - w[i]
            path[i] = 1
            maxS_ += w[i]
    print(path,maxS_==maxS)
```

```python
def backpack01_optim(w,n,c):
    '''
    因为每次只需要根据上一层信息做状态转移，因此只需要一维数组即可
    '''
    curc = [False for j in range(c+1)]
    # 第0个物品装或不装
    curc[0] = True
    if c >= w[0]:
        curc[w[0]] = True
    for i in range(1,n):
        # 装或不装，装不需要进行状态转移了，因为当前已经是True。
        for j in range(c,-1,-1):
            if j+w[i] <= c and curc[j]:curc[j+w[i]]=True
    for k in range(c,-1,-1):
        if curc[k]:break
    return (k,curc)
```

# 解题思路
对于每个物品来说，都有两种选择，装进背包或者不装进背包。对于 n 个物品来说，总的装法就有 2^n 种，去掉总重量超过 Wkg 的，从剩下的装法中选择总重量最接近 Wkg 的。不过，我们如何才能不重复地穷举出这 2^n 种装法呢？这里就可以用回溯的方法。我们可以把物品依次排列，整个问题就分解为了 n 个阶段，每个阶段对应一个物品怎么选择。先对第一个物品进行处理，选择装进去或者不装进去，然后再递归地处理剩下的物品。
